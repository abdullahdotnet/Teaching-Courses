# Office management system

**Diagram link**: https://miro.com/app/board/uXjVN3T3XfI=/?share_link_id=620526843681

**Hierarchy**: There will be a CEO. It will have some departments in its possession. A manager will head a department. Each department will have some teams inside it. A team lead will head a team. There will be a number of employees in the team.  

# Classes
## 1. Employee
### Attributes:
`id` is a unique identifier for the employee. Autoincrement id whenever the employee is created.    
`name` name of the employee.   
`team` team class object represents in which team, the employee is associated  
`contact_info` is a dictionary that can include email and phone number  
`salary` shows the annual salary of the employee  
`role` represents the position of the employee in the company  
`tasks` list of tasks assigned to the employee  
### Methods:    
`__repr__()` returns name and role as string  
`update_info(self, contact_info)` new contact_info as argument  
`increment_pay(self, perc)` give the percentage of salary increment  
`display_tasks(self)` display all tasks with the numbering, while each task on a separate line  
`complete_task(self, task_index)` get the index number of the task and remove from task list  


## 2. TeamLead(Employee)
### Attributes:
`no_of_projects` Number of projects  
`team` team object which the team lead supervises     
### Methods:
`assign_task(self, task, employee)` adds task to the tasks list of the employee object  
`motivate_team(self)` prints the motivation message for the team.  
`get_employees(self)` returns the list of employees from the team object  
`add_employee(self, employee)` add the employee into the team object's employee list  
`remove_employee(self, employee)` remove the employee from the team object's employee list  

## 3. Manager(Employee)
### Attributes:
`department` department object which the manager lead supervises     
### Methods:
`add_team(self, team)` add team in the department object  
`remove_team(self, team)` remove team from the department object
`get_teams(self)` returns the team list from the department object  
`assign_team_lead(self, employee, team)` creates a team-lead object from attributes of employee object and assign the team-lead object as an attribute in the team. Also delete the previous employee object using the `del` keyword to avoid any contradiction.

## 4. Team
### Attribute
`id` is a unique identifier for the team.  
`name` name of the team.   
`team_lead` team-lead object that supervises this team  
`department` refers to the department object with which the team is associated  
`employees` list of employee objects included in the team   
### Methods:   
`__repr__()` returns the name of the team, its team_lead's name and department name   
`__eq__()` checks whether the two teams are same or not. (Use id to identify)  

## 5. Department
### Attribute
`id` is a unique identifier for the team.  
`name` name of the team.   
`manager` manager object that supervises this department  
`budget` total amount allocated to the department annually  
`teams` list of team objects included in the department   
### Methods:   
`__repr__()` returns the name of the department and its manager's name   
`__eq__()` checks whether the two departments are same or not. (Use id to identify)
